***Q2. Included is the 1-year sales data of customers with their corresponding sales and the number of transactions they have completed in the past year.***

***a. Read the csv file and create a dataframe***

```python
import numpy as np
```

```python
df1 = pd.read_csv('/content/dataset.csv')
```

```python
df1.head()
```

***b. Create a calculated column of `Average Order Value (AOV)` for the customers***

```python
df1['AOV'] = df1['Sales_Value']/df1['Transaction_count']
```

```python
df1.head()
```

c. Classify customers as:

 AOV: <=30 or >30

 Transaction count: {1,2,3+}

```python
# Classify and segment customers
conditions = [
    (df1['AOV'] <= 30) & (df1['Transaction_count'] == 1),
    (df1['AOV'] <= 30) & (df1['Transaction_count'] == 2),
    (df1['AOV'] <= 30) & (df1['Transaction_count'] >= 3),
    (df1['AOV'] > 30) & (df1['Transaction_count'] <= 2),
    (df1['AOV'] > 30) & (df1['Transaction_count'] >= 3),
]
```

```python
segments = ['One Off', 'Regular', 'Frequent', 'High Value', 'Loyal']
df1['Customer_Segment'] = np.select(conditions, segments, default='Unknown')
```

***e. Summarize the customer segments with customer count, transaction count and their sales value***

```python
df1.head()
```

```python
summary = df1.groupby('Customer_Segment').agg(
    Customer_Count=('Customer_ID', 'nunique'),
    Transaction_Count=('Transaction_count', 'sum'),
    Sales_Value=('Sales_Value', 'sum')
).reset_index()
```
